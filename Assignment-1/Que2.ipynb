### Que2 - Answer

*Key idea* -

First we will prove that the given objective cost function is related to the k-means objective cost function. After this, we can apply k-means or k-means++ algorithm.

*Proof* -

Suppose we have k-clusters denoted by set $C = \{ C_{1}, C_{2},...C_{k} \} $, and each data point is a vector of $d\times1$ dimension, then the given objective function is -

$cost(C) =   \sum_{i} \frac{1}{|C_{i}|} \sum_{x,y \in C_{i}} \|x-y\|_{2}^{2} = \sum_{i} \frac{1}{|C_{i}|} \sum_{y \in C_{i}} \sum_{x \in C_{i}} \|x-y\|_{2}^{2}  $ 

Now, for a fixed $C_{i}$, let us fix a $y \in C_{i}$

Suppose $ m_{i} = \frac{1}{|C_{i}|} \sum_{x \in C_{i}} x $ i.e. $m_{i}$ ( a vector of $d\times1$ dimension)  is the mean of all points in the cluster $C_{i}$.

$ \sum_{x \in C_{i}} \|x-y\|_{2}^{2} =  \sum_{x \in C_{i}} \|x-m_{i}+m_{i}-y\|_{2}^{2} $

$ =  \sum_{x \in C_{i}} \|x-m_{i}\|_{2}^{2} + \sum_{x \in C_{i}} \|y-m_{i}\|_{2}^{2} + \sum_{x \in C_{i}} 2(m_{i} - y)^{T}(x-m_{i}) $


$ =  \sum_{x \in C_{i}} \|x-m_{i}\|_{2}^{2} + |C_{i}| \|y-m_{i}\|_{2}^{2} + 2(m_{i} - y)^{T} \sum_{x \in C_{i}} (x-m_{i}) $


$ =  \sum_{x \in C_{i}} \|x-m_{i}\|_{2}^{2} + |C_{i}| \|y-m_{i}\|_{2}^{2} + 2(m_{i} - y)^{T} ( \sum_{x \in C_{i}} x- \sum_{x \in C_{i}} m_{i} ) $


$ =  \sum_{x \in C_{i}} \|x-m_{i}\|_{2}^{2} + |C_{i}| \|y-m_{i}\|_{2}^{2} + 2(m_{i} - y)^{T} ( \sum_{x \in C_{i}} x- |C_{i}| m_{i} ) $

(But, by definition of $m_{i}$, $\sum_{x \in C_{i}} x- |C_{i}| m_{i} = (0, 0, ... d-times)^{T} $ i.e. it is a zero vector of $d\times1$ dimension)

$ =  \sum_{x \in C_{i}} \|x-m_{i}\|_{2}^{2} + |C_{i}| \|y-m_{i}\|_{2}^{2} + 0 $


$ =  \sum_{x \in C_{i}} \|x-m_{i}\|_{2}^{2} + |C_{i}| \|y-m_{i}\|_{2}^{2} $


Hence, the cost now becomes,


$cost(C) = \sum_{i} \frac{1}{|C_{i}|} \sum_{y \in C_{i}} ( \sum_{x \in C_{i}}  \|x-m_{i}\|_{2}^{2} + |C_{i}| \|y-m_{i}\|_{2}^{2}  ) $

$ = \sum_{i} \frac{1}{|C_{i}|} (   \sum_{y \in C_{i}} \sum_{x \in C_{i}}  \|x-m_{i}\|_{2}^{2} + \sum_{y \in C_{i}} |C_{i}| \|y-m_{i}\|_{2}^{2}  ) $

$ = \sum_{i} \frac{1}{|C_{i}|} ( |C_{i}| \sum_{x \in C_{i}}  \|x-m_{i}\|_{2}^{2} + \sum_{y \in C_{i}} |C_{i}| \|y-m_{i}\|_{2}^{2}  ) $

$ = \sum_{i} \frac{1}{|C_{i}|} ( |C_{i}| \sum_{x \in C_{i}}  \|x-m_{i}\|_{2}^{2} + |C_{i}| \sum_{y \in C_{i}}  \|y-m_{i}\|_{2}^{2}  ) $

But, these two terms are actually the same.

$cost(C) = \sum_{i} \frac{1}{|C_{i}|} ( 2|C_{i}| \sum_{x \in C_{i}}  \|x-m_{i}\|_{2}^{2} ) $


$ =  2\sum_{i} \sum_{x \in C_{i}}  \|x-m_{i}\|_{2}^{2} $


Where, $ m_{i} = \frac{1}{|C_{i}|} \sum_{x \in C_{i}} x $

This is actually two times of the k-means objective function if we assume that each point has been assigned to a cluster where it's distance from the cluster mean $m_{i}$ is least among other cluster.

Hence, to find such clustering $C = \{ C_{1},C_{2},...C_{k} \} $ we can apply *k-means++ algorithm* (Llyod's algorithm in particular) on the given points because the given cost turns out into k-means cost with a constant multiplication factor (which does not matter).
